## Carregamento dos Dados


**Dicionario de dados**
*   https://github.com/sthemonica/alura-voz


```python
alura_voice_raw = pd.read_json('https://raw.githubusercontent.com/sthemonica/alura-voz/main/Dados/Telco-Customer-Churn.json')
```

```python
alura_voice_raw.head()
```

# Aplicar algumas transformações sobre os dados

```python
customers_df = pd.DataFrame.from_records(alura_voice_raw['customer'])
phone_df = pd.DataFrame.from_records(alura_voice_raw['phone'])
internet_df = pd.DataFrame.from_records(alura_voice_raw['internet'])
account_df = pd.DataFrame.from_records(alura_voice_raw['account'])
charges_df = pd.DataFrame.from_records(account_df['Charges'])
```

```python
alura_voice =  pd.concat([alura_voice_raw[['customerID', 'Churn']],customers_df, phone_df, internet_df, account_df[['Contract', 'PaperlessBilling', 'PaymentMethod']], charges_df], axis=1)
alura_voice['Churn'] = alura_voice['Churn'].map({"Yes": "Sim", "No":"Nao"})
```

```python
alura_voice.head()
```

## Colunas e valores

```python
for i in alura_voice.columns:
  if(alura_voice[i].nunique() < 10):
    print (i, alura_voice[i].nunique())
    print (alura_voice[i].value_counts())
    print('-----------')
```
